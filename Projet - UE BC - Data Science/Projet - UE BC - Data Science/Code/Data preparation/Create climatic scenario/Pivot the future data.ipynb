```python
import pandas as pd
import os
from scipy.stats import zscore
import matplotlib.pyplot as plt
```

```python
df_meteo  = pd.read_csv('Scenario_Climatique.csv')
df_meteo.head()
```

```python
df_meteo = df_meteo.drop(df_meteo.columns[0], axis=1)
df_meteo
```

```python
# Convert the date column to datetime format
df_meteo['Time'] = pd.to_datetime(df_meteo['Time'])

# Extract, month, and year columns
df_meteo['Month'] = df_meteo['Time'].dt.month
df_meteo['Year'] = df_meteo['Time'].dt.year
df_meteo = df_meteo.drop(df_meteo.columns[1], axis=1)
column_to_move = df_meteo.pop("Month")
df_meteo.insert(1, 'Month', column_to_move)
column_to_move = df_meteo.pop("Year")
df_meteo.insert(2, 'Year', column_to_move)
df_meteo
```

```python
L = list(df_meteo.columns)

elements_to_remove = ['Year', 'Month', 'County']

L = [item for item in L if item not in elements_to_remove]
```

```python
df_meteo1 = df_meteo[df_meteo['County'] == 'BOTTINEAU']
months = ['January', 'February', 'March', 'April', 'May', 'June',
          'July', 'August', 'September', 'October', 'November', 'December']
monthly_dfs = {}
columns_to_expand = L
# Pivot the table for each month
for i,month in enumerate(months):
    # Filter the DataFrame for the current month
    month_df = df_meteo1[df_meteo1['Month'] == i+1 ]

    # Rename the columns to include the month
    renamed_columns = {col: f'{col} {month}' for col in columns_to_expand}
    month_df = month_df.rename(columns=renamed_columns)

    # Drop the 'Month' column and keep only the expanded columns
    month_df = month_df[['Year'] + list(renamed_columns.values())]

    monthly_dfs[month] = month_df

# Merge all monthly DataFrames
from functools import reduce
expanded_df1 = reduce(lambda left, right: pd.merge(left, right, on=['Year'], how='outer'),
                     monthly_dfs.values())
expanded_df1['County'] = 'BOTTINEAU'
column_to_move = expanded_df1.pop('County')
expanded_df1.insert(1, 'County', column_to_move)
expanded_df1
```

```python
df_meteo2 = df_meteo[df_meteo['County'] == 'WELLS']
months = ['January', 'February', 'March', 'April', 'May', 'June',
          'July', 'August', 'September', 'October', 'November', 'December']
monthly_dfs = {}
columns_to_expand = L
# Pivot the table for each month
for i,month in enumerate(months):
    # Filter the DataFrame for the current month
    month_df = df_meteo2[df_meteo2['Month'] == i+1 ]

    # Rename the columns to include the month
    renamed_columns = {col: f'{col} {month}' for col in columns_to_expand}
    month_df = month_df.rename(columns=renamed_columns)

    # Drop the 'Month' column and keep only the expanded columns
    month_df = month_df[['Year'] + list(renamed_columns.values())]

    monthly_dfs[month] = month_df

# Merge all monthly DataFrames
from functools import reduce
expanded_df2 = reduce(lambda left, right: pd.merge(left, right, on=['Year'], how='outer'),
                     monthly_dfs.values())
expanded_df2['County'] = 'WELLS'
column_to_move = expanded_df2.pop('County')
expanded_df2.insert(1, 'County', column_to_move)
expanded_df2
```

```python
df_meteo3 = df_meteo[df_meteo['County'] == 'CASS']
months = ['January', 'February', 'March', 'April', 'May', 'June',
          'July', 'August', 'September', 'October', 'November', 'December']
monthly_dfs = {}
columns_to_expand = L
# Pivot the table for each month
for i,month in enumerate(months):
    # Filter the DataFrame for the current month
    month_df = df_meteo3[df_meteo3['Month'] == i+1 ]

    # Rename the columns to include the month
    renamed_columns = {col: f'{col} {month}' for col in columns_to_expand}
    month_df = month_df.rename(columns=renamed_columns)

    # Drop the 'Month' column and keep only the expanded columns
    month_df = month_df[['Year'] + list(renamed_columns.values())]

    monthly_dfs[month] = month_df

# Merge all monthly DataFrames
from functools import reduce
expanded_df3 = reduce(lambda left, right: pd.merge(left, right, on=['Year'], how='outer'),
                     monthly_dfs.values())
expanded_df3['County'] = 'CASS'
column_to_move = expanded_df3.pop('County')
expanded_df3.insert(1, 'County', column_to_move)
expanded_df3
```

```python
df = pd.concat([expanded_df1, expanded_df2, expanded_df3], ignore_index = True)
df
```
