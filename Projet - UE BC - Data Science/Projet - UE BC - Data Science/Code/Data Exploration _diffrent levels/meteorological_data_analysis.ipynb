```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

# Define the start and end years for the analysis
start_year = 1991
end_year = 2022

# Choose a specific 'Type' and 'Variable'
selected_type = 'WHEAT'
selected_variable = 'YIELD, MEASURED IN BU / ACRE'

# Filter the DataFrame for the selected 'Type' and 'Variable' within the specified years
filtered_df = merged_df[(merged_df['Type'] == selected_type) & (merged_df['Variable'] == selected_variable) & (merged_df['Year'] >= start_year) & (merged_df['Year'] <= end_year)]

# Select the meteorological data columns
meteorological_columns = [
    'Max Temp', 'Min Temp', 'Avg Temp', 'Avg Bare Soil Temp', 'Avg Turf Soil Temp',
    'Avg Wind Speed', 'Max Wind Speed', 'Total Solar Rad', 'Avg Penman PET',
    'Total Penman PET', 'Total Rainfall', 'Avg Dew Point', 'Avg Wind Chill'
]

# Normalize the 'Value' column using Min-Max scaling
scaler = MinMaxScaler()
filtered_df['Normalized Value'] = scaler.fit_transform(filtered_df[['Value']])

# Calculate the correlation matrix for each agricultural district
correlation_matrices = {}
for district, data in filtered_df.groupby('Ag District'):
    test = data[['Normalized Value'] + meteorological_columns]
    print(district, ' ', test.shape)
    correlation_matrix = data[['Normalized Value'] + meteorological_columns].corr()
    correlation_matrices[district] = correlation_matrix

# Create subplots for each district's correlation heatmap
n_districts = len(correlation_matrices)
fig, axes = plt.subplots(1, n_districts, figsize=(10 * n_districts, 10))
fig.suptitle(f'Correlation Heatmaps for {selected_variable} and Meteorological Data (Years {start_year}-{end_year}) by Agricultural District')

# Plot each district's heatmap
for i, (district, correlation_matrix) in enumerate(correlation_matrices.items()):
    ax = axes[i]
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", ax=ax)
    ax.set_title(f'District {district}')
    ax.set_xlabel('Meteorological Variable')
    ax.set_ylabel('Meteorological Variable')

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from scipy.stats import pearsonr

# Define the start and end years for the analysis
start_year = 1991
end_year = 2000

# Choose a specific 'Type' and 'Variable'
selected_type = 'WHEAT'
selected_variable = 'YIELD, MEASURED IN BU / ACRE'

# Filter the DataFrame for the selected 'Type' and 'Variable' within the specified years
filtered_df = merged_df[(merged_df['Type'] == selected_type) & (merged_df['Variable'] == selected_variable) & (merged_df['Year'] >= start_year) & (merged_df['Year'] <= end_year)]

# Select the meteorological data columns
meteorological_columns = [
    'Max Temp', 'Min Temp', 'Avg Temp', 'Avg Bare Soil Temp', 'Avg Turf Soil Temp',
    'Avg Wind Speed', 'Max Wind Speed', 'Total Solar Rad', 'Avg Penman PET',
    'Total Penman PET', 'Total Rainfall', 'Avg Dew Point', 'Avg Wind Chill'
]

# Normalize the 'Value' column using Min-Max scaling
scaler = MinMaxScaler()
filtered_df['Normalized Value'] = scaler.fit_transform(filtered_df[['Value']])

# Create a dictionary to store correlation matrices and p-values for each district
correlation_data = {}

# Calculate the correlation matrix and p-value for each agricultural district
for district, data in filtered_df.groupby('Ag District'):
    correlation_matrix = data[meteorological_columns].corr()
    p_value_matrix = data[meteorological_columns].apply(lambda x: data.corrwith(x)).fillna(1)
    correlation_data[district] = {'correlation_matrix': correlation_matrix, 'p_value_matrix': p_value_matrix}

# Create subplots for each district's correlation heatmap
n_districts = len(correlation_data)
fig, axes = plt.subplots(1, n_districts, figsize=(5 * n_districts, 5))
fig.suptitle(f'Correlation Heatmaps for {selected_variable} and Meteorological Data (Years {start_year}-{end_year}) by Agricultural District')

# Plot each district's heatmap with p-values
for i, (district, data) in enumerate(correlation_data.items()):
    ax = axes[i]
    sns.heatmap(data['correlation_matrix'], annot=True, cmap='coolwarm', fmt=".2f", cbar=False, ax=ax)
    ax.set_title(f'District {district}')
    ax.set_xlabel('Meteorological Variable')
    ax.set_ylabel('Meteorological Variable')

    # Add p-values as annotations
    for i in range(len(meteorological_columns)):
        for j in range(len(meteorological_columns)):
            ax.text(j + 0.5, i + 0.5, f'p={data["p_value_matrix"].iloc[i, j]:.2f}', ha='center', va='center', fontsize=8, color='red')

plt.tight_layout()
plt.show()
```
